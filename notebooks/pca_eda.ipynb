```python
import pandas as pd
import matplotlib.pyplot as plt
import umap
```

```python
df = pd.read_csv('../data/options.csv', nrows=400000)
```

```python
df.drop(columns=['Unnamed: 0', 'symbol', 'date'], inplace=True)
X = df.drop(columns=['DITM_IV'])
y = df['DITM_IV']
```

```python
del df
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(X)
```

```python
plt.scatter(embedding[:, 0], embedding[:, 1], c=y)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
reduced_x = pca.fit_transform(X)
```

```python
len(reduced_x[:,0])
```

```python
plt.scatter(reduced_x[:, 0], reduced_x[:, 1], c=y)
```
